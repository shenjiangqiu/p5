```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Load Data
college_data = pd.read_csv("./p5-data/college_data.csv")
#college_data.head()
```

```python
cdata = college_data.iloc[:,3:]
cdata.head()
cdata.columns
```

```python
#Standarizing the features
scdata = StandardScaler().fit_transform(cdata.values)
cdataScaled = pd.DataFrame(data= scdata, columns=cdata.columns)
cdataScaled.head()
```

```python
#PCA
pca = PCA()
pca.fit(cdataScaled)

print("principal components")
print(pca.components_)
print("\nsingular values")
print(pca.singular_values_)
print("\nexplained variance ratio")
print(pca.explained_variance_ratio_)
```

```python
pc = pd.DataFrame(pca.components_)
pc.columns = cdata.columns
pc.index.name = "PC"
pc
```

## b

```python
# PC scores
#pca = PCA(n_components=2)
X = pca.fit_transform(cdataScaled)
Xnew = pd.DataFrame(X)
Xnew.head()
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2');
```

```python
# Visualize the principal component directions
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1),
                    color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i],
                    color = 'g', ha = 'center', va = 'center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()
```

```python
myplot(X[:,0:2], np.transpose(pca.components_[0:2,:]), cdata.columns[0:2])
```

## C

```python
#Explianing Variance
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("Principal Components")
plt.ylabel("Proportion of Variance Explained");
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel("Principal Components")
plt.ylabel("Cumulative Proportion of Variance Explained")
plt.title('Cumulative Proportion of Variance Explained for Pizza Data');
```
